## SALBP-1

The last six columns describe the minimum number of stations after `x` seconds, where `x` is the column name.

```python
df_salbp1 = pd.read_excel(
    str(Path("../data/raw/albp-datasets/SALBP-1993/SALBP data sets.xlsx")),
    sheet_name="SALBP-1 data set",
    skiprows=24,
    skipfooter=4
).dropna(axis=1)

df_salbp1
```

```python
df = pd.merge(precedence_graphs, df_salbp1, left_on="Name", right_on="Precedence graph")

def write_salbp1(graph: str, N: int, c: int, OS: float):
    data = Path("..", "data")
    raw = Path(data, "raw")
    processed = Path(data, "processed")
    graph = Path(graph)
    graph_path = Path(raw, "albp-datasets", "SALBP-1993", graph)
    with open(graph_path, "r") as f:
        data = f.read().split("\n")
    t = [i for i in data[1:N + 1]]
    P = [i for i in data[N + 1:] if (i != "-1,-1") and (len(i.split(",")) > 1)]

    problem = "\n".join([
        SECTION_NUM_TASKS, str(N), "",
        SECTION_CYCLE_TIME, str(c), "",
        SECTION_ORDER_STRENGTH, str(OS), "",
        SECTION_TASK_TIMES, *t, "",
        SECTION_PRECEDENCES, *P
    ])
    destination = Path(processed, "albp-datasets", "SALBP-1993")
    destination.mkdir(parents=True, exist_ok=True)
    with open(Path(destination, graph.stem + f"_{c}.alb"), "w") as f:
        f.write(problem)

for i, inst in df.iterrows():
    write_salbp1(inst['File name'], inst['n'], inst['c'], inst['OS'])
```
